# Define a function that will perform pre processing 
1. Stop Word removal
2. Remove Special Characters and unwanted spaces 
3. Punctuation removal
4. Convert all characters to lower case 

# ** Disclosure: I am re-using this function from code snippet found on stack overflow at the following link*
https://stackoverflow.com/questions/54396405/how-can-i-preprocess-nlp-text-lowercase-remove-special-characters-remove-numb/54398984 

```
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.corpus import stopwords
import re
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer() 

def preprocess(sentence):
    sentence=str(sentence)
    sentence = sentence.lower()
    sentence=sentence.replace('{html}',"") 
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', sentence)
    rem_url=re.sub(r'http\S+', '',cleantext)
    rem_num = re.sub('[0-9]+', '', rem_url)
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(rem_num)  
    filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords.words('english')]
    stem_words=[stemmer.stem(w) for w in filtered_words]
    lemma_words=[lemmatizer.lemmatize(w) for w in stem_words]
    return " ".join(filtered_words)
```

```
df['text']=df['text'].map(lambda s:preprocess(s)) 

```

# Vectorize the Features

```
from sklearn.feature_extraction.text import CountVectorizer
```

```
cvect = CountVectorizer(ngram_range=(1,2))
```

```
# create document-term matrices
x_train_ct = cvect.fit_transform(x_train)
x_test_ct = cvect.transform(x_test)
```
